```python
import pandas as pd
data =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/unprocessed.csv")
data
```

```python
data['Index']=data.index
data
```

```python
data['Gender_Replace']=data['Gender'].replace({'M':'Male','F':'Female'})
data
```

```python
data['Gender_Numeric']=data['Gender_Replace'].map({'Male':1,'Female':0})
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data[['Standardized_Age','Standardized_Income']]=scaler.fit_transform(data[['Age','Income']])
data
```

```python
import matplotlib.pyplot as plt
data['Income']=pd.to_numeric(data['Income'],errors='coerce') #converting to numeric,coercing errors to NaN
data=data.dropna(subset=['Income']) #Drop rows with Nan in Income
#Plotting
plt.boxplot(data['Income'])
plt.title('Income Boxplpot - Identifying Outliers')
plt.show()
```
